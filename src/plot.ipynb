```python
# I recommend opening this in VSCode if you're not familiar with Jupyter notebooks
# Create and activate a virtual environment, on a mac this would be:
#
# python3 -m venv venv
# source venv/bin/activate

# Then install the dependencies:
# pip install ipykernel matplotlib numpy

# Select your virtual environment as the kernel
# Then hit the "Run All" button at the top

# If you want to keep it pretty, I formatted it with black:
# pip install black "black[jupyter]"
# black ./src/plot.ipynb


import numpy as np
import matplotlib.pyplot as plt

# Sample power curve data
power_curve = [
    500,
    499,
    498,
    *[490] * 5,
    *[450] * 20,
    *[400] * 60,
    *[350] * 120,
    *[300] * 180,
    *[250] * 240,
    *[200] * 300,
    *[150] * 360,
    *[100] * 500,
    *[50] * 1000,
    *[40] * 2000,
    *[35] * 2000,
]

# for the x axis - not sure this is necessary
time_seconds = np.arange(len(power_curve))

plt.plot(time_seconds, power_curve)

# Set x-axis to log scale
plt.xscale("log")

# Set x-axis ticks and labels
plt.xticks(
    [2, 5, 10, 20, 30, 60, 120, 300, 600, 1200, 2400, 3600, 7200, 14400],
    [
        "2s",
        "5s",
        "10s",
        "20s",
        "30s",
        "1m",
        "2m",
        "5m",
        "10m",
        "20m",
        "40m",
        "1h",
        "2h",
        "4h",
    ],
)

plt.xlabel("Duration")
plt.ylabel("Power")
plt.title("Power Curve")

plt.show()
```
